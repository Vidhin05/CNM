```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from scipy.interpolate import lagrange as l
from sympy import *

def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
    
def lag(x, y, xp, laglab):
    poly = l(x, y)
    plt.plot(xp, poly(xp), label=laglab)
    plt.grid(True)
    plt.legend(loc='best')
    print("The Lagrangian polynomial is:")
    print(poly)
    
def lagwithfunc(f, x, y, xp, lab):
    poly = l(x, y)
    plt.plot(xp, f(xp), label = lab)
    plt.plot(xp, poly(xp), label="Lagrange Polynomial")
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
    print("The Lagrangian polynomial is:")
    print(poly)
```

```python
f = lambda x: x**0.5
lagwithfunc(f, np.array([1, 4]), np.array([1, 2]), np.arange(0,5,0.01), "y=x^0.5")
```

```python
xp = np.arange(0, 1.5, 0.01)
lag(np.array([0.82, 0.83]), np.array([2.270500, 2.293319]), np.arange(0, 1.5, 0.01), "First order Lagrange Polynomial")
plt.plot(xp, np.exp(xp), label="y = e^x")
lag(np.array([0.82, 0.83, 0.84]), np.array([2.270500, 2.293319,2.316367]), np.arange(0, 1.5, 0.01), "Second order Lagrange Polynomial")
plt.show()
```

```python
lag(np.array([0, 1, 2]), np.array([-1, -1, 7]), np.arange(0,5,0.01), "Lagrange Polynomial")
plt.show()
```
