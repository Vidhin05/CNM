```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
```

```python
a = 1.0000
b = 2.0000
c = (a+b)/2
def f(a):
    return ((a**6) -(a +1))
df = pd.DataFrame( columns = ['a','b','c = (a+b)/2','f(c)','b-c=c-a','Sign of f(a)f(c)','Assign c='])
i = 0
plot(np.arange(-2,2,0.001), f(np.arange(-2,2,0.001)), "x^6 - x -1")
while ((b-a)>0.0001) :
    flag = 0
    if f(a)*f(c) >0:
        flag=1
        a = c
        c = (a+b)/2
    else:
        b = c
        c = (a+b)/2
    i=i+1        
    if(flag==1):
        x='a'
        sign="Positive"
    else: 
        x='b'
        sign="Negative"
    data = [a,b,c,f(c),b-c,sign,x]
    df.loc[i] = data
print(str.format('{0:.4f}', c))
plot([k for k in range(i)], df['c = (a+b)/2'],"Convergence towards root")
df
```
