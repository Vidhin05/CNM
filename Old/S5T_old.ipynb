```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize
from sympy import *

def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
    
def NewtonRhapson(b):
  x = Symbol('x')  
  y = f(x)
  yprime = diff(y,x)
  root = optimize.newton(f, b)
  return root    
    
def NewtonRhapsone(b):
  x = Symbol('x')  
  y = f(x)
  yprime = diff(y,x)
  root = optimize.newton(f, b,maxiter=1,tol=100)
  return root
```

```
def f(x):
      return (x**3 - x**2 -x -1)
x = np.arange(-3,3,0.001)
y = x**3 - x**2 -x -1
plot(x,y,"x**3 - x**2 -x -1")

start=[1.5]
for i in range(6):
  start.append(NewtonRhapsone(start[i]))
plot([k for k in range(7)],start,"Convergence towards root")
start
```
