```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize
from sympy import *

def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
    
def NewtonRhapson(b):
  x = Symbol('x')  
  y = f(x)
  yprime = diff(y,x)
  root = optimize.newton(f, b, fprime= lambdify(x, yprime, 'numpy'))
  return root    
    
def NewtonRhapsone(b):
  x = Symbol('x')  
  y = f(x)
  yprime = diff(y,x)
  root = optimize.newton(f, b, fprime= lambdify(x, yprime, 'numpy'),maxiter=1,tol=10000)
  return root

def iterate(init):  
  start=[init-0.001, init]
  i = 1
  while(abs(start[i]-start[i-1])>0.0001):
    i = i+1
    start.append(NewtonRhapsone(start[i-1]))   
  plot([k for k in range(i)],start[1:],"Convergence towards root from " + str(init))
  print(start[1:],sep="\n")
  return i
```

```
def f(x):
      return (x**6 - x -1)
x = np.arange(-2,2,0.001)
plot(x,(x**6 - x -1),"x^6 - x -1")
plot(x, 6*(x**5)-1,"6x^5 -1")

iterate(1.5)
```
