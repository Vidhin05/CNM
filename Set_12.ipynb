```python
% matplotlib inline
import numpy as np
from   matplotlib import pyplot as plt
```

# Q1. a) 

## $y′(x) = −y(x)$ 
## $y(0) = 1$

```python
x0 = 0
y0 = 1
xf = 6
k = 1
plt.figure(figsize= (25,12))
for h in [0.2, 0.1, 0.05]:
    n = int(6/h+1)
    x = np.linspace(x0, xf + h, n)
    y = np.zeros([n])
    y[0] = y0

    for i in range(1, n) :
        y[i] = h*(-y[i-1]) + y[i-1]
    
    plt.subplot(2, 3, k)
    plt.plot(x, y, '.')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("h = "+ str(h))
    plt.plot(x, np.exp(-x))
    
    plt.subplot(2, 3, k + 3)
    plt.plot(x, y - np.exp(-x))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("Error for h = "+ str(h))
    k += 1
```

# Q1. b) 

## $y′(x)=[y(x)+x^2−2]/(x+1)$
## $y(0) = 2$

```python
x0 = 0
y0 = 2
xf = 6
k = 1
plt.figure(figsize= (25,12))
for h in [0.2, 0.1, 0.05]:
    n = int(6/h + 1)
    x = np.linspace(x0, xf + h, n)
    y = np.zeros([n])
    y[0] = y0

    for i in range(1, n) :
        y[i] = h *((y[i-1] + (x[i-1])**2 - 2)/(x[i-1] + 1)) + y[i-1]
    
    plt.subplot(2, 3, k)
    plt.plot(x, y, '.')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("h = "+ str(h))
    plt.plot(x, (x**2 + 2*x + 2) - (2*(x+1)*np.log(x+1)))
    
    plt.subplot(2, 3, k + 3)
    plt.plot(x, y - ((x**2 + 2*x + 2) - (2*(x+1)*np.log(x+1))))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("Error for h = "+ str(h))
    k += 1
```
