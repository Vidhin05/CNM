```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize
from sympy import *

def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
    
def NewtonRhapson(b):
  x = Symbol('x')  
  y = f(x)
  yprime = diff(y,x)  
  yprime2 = diff(yprime, x)
  root = optimize.newton(f, b, fprime= lambdify(x, yprime, 'numpy'),fprime2=lambdify(x, yprime2, 'numpy'))
  print(root)
  
 
```

```
def f(x):
      return (x**3 - x**2 -x -1)
x = np.arange(-3,3,0.001)
y = x**3 - x**2 -x -1
plot(x,y,"x**3 - x**2 -x -1")

NewtonRhapson(1.5)
```

```
def f(x):
      return (1+ 0.3 * np.cos(x) - x)
x = np.arange(-2,2,0.001)
y = 1+ 0.3 * np.cos(x) - x
plot(x,y,"1+ 0.3 * cos(x) - x")
root = optimize.newton(f, 1, fprime=lambda x: ((-0.3 * np.cos(x)) - 1), fprime2=lambda x: 0.3 * np.sin(x))
print(root)
```

```
def f(x):
      return (0.5 - np.cos(x) +np.sin(x))
x = np.arange(-2,2,0.001)
y = 0.5 - np.cos(x) +np.sin(x)
plot(x,y,"0.5 - cos(x) +sin(x)")
root = optimize.newton(f, 0.5, fprime=lambda x: (np.sin(x)+np.cos(x)), fprime2=lambda x: (np.cos(x)-np.sin(x)))
print(root)
```

```
def f(x):
      return (np.exp(-x) -(x))
x = np.arange(-2,2,0.001)
y = np.exp(-1*x) -(x)
plot(x,y,"e^(-x)-x")
root = optimize.newton(f, 0.5, fprime=lambda x: (-np.exp(-x) -(1)), fprime2=lambda x: (np.exp(-x)))
print(root)
```

```
def f(x):
      return (np.exp(-x) -np.sin(x))
x = np.arange(0,5,0.001)
y = np.exp(-1*x) -np.sin(x)
plot(x,y,"e^(-x)-sin(x)")
root = optimize.newton(f, 0.5, fprime=lambda x: (-np.exp(-x) -(np.cos(x))), fprime2=lambda x: (np.exp(-x)+np.sin(x)))
print(root)
root = optimize.newton(f, 3, fprime=lambda x: (-np.exp(-x) -(np.cos(x))), fprime2=lambda x: (np.exp(-x)+np.sin(x)))
print(root)
```

```
def f(x):
      return (1 - x +np.sin(x))
x = np.arange(-1,3,0.001)
y = 1 - x +np.sin(x)
plot(x,y,"1 - x + sin(x)")
root = optimize.newton(f, 1, fprime=lambda x: (-1 +(np.cos(x))), fprime2=lambda x: (-np.sin(x)))
print(root)
```

```
def f(x):
      return (np.tan(x) - x)
x = np.arange(-10,10,0.1)
y = np.tan(x) - x 
plot(x,y,"tan(x) - x")
root = optimize.newton(f, 4.5, fprime=lambda x: ((1/(np.cos(x))**2)-1), fprime2=lambda x: ((2*(1/(np.cos(x))**2))*np.tan(x)))
print(root)
```
