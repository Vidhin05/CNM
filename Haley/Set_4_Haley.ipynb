```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize
from sympy import *

def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
    
def NewtonRhapson(b):
  x = Symbol('x')  
  y = f(x)
  yprime = diff(y,x)  
  yprime2 = diff(yprime, x)
  root = optimize.newton(f, b, fprime= lambdify(x, yprime, 'numpy'),fprime2=lambdify(x, yprime2, 'numpy'))
  print(root)
```

```
def f(x):
      return (x**6 - x -1)
x = np.arange(-2,2,0.001)
plot(x,(x**6 - x -1),"x^6 - x -1")
NewtonRhapson(1.5)
NewtonRhapson(-1.5)
```
