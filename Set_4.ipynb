```python
import numpy  as np
import pandas as pd
import math
import matplotlib.pyplot as plt

def plot(x, y,lab):
    plt.plot(x, y, label=lab)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
```

```python
table = pd.DataFrame(columns=['n', 'x(n)', 'f(x)', "f'(x)", 'x(n) - x(n-1)', 'x(n+1)', 'x(n+1) - x(n)'])

n    = 0
x_n  = 1.0
f_x  = x_n**6 -x_n -1
dfdx = 6*(x_n**5) - 1
dif1 = 0
x_n1 = x_n - (f_x/dfdx)
dif2 = x_n1 - x_n
table.loc[0] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]

while(1):
  n += 1
  dif1 = x_n1 - x_n
  x_n = x_n1
  f_x  = x_n**6 -x_n -1
  dfdx = 6*(x_n**5) - 1
  x_n1 = x_n - (f_x/dfdx)
  dif2 = x_n1 - x_n
  table.loc[n] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]
  if(abs(dif1) < 0.0001):
    break
    
x = np.arange(-3,3,0.001)
y = x**3 - x**2 -x -1
plot(x,y,"x**3 - x**2 -x -1")
x = np.arange(-3,3,0.001)
y = 3*x**2 - 2*x -1
plot(x,y,"3x^2 - 2x -1")

plt.plot(table['n'], table['x(n+1)'])
plt.title("Convergence of root between 1 and 2")
plt.show()
plt.plot(table["n"], table["f'(x)"])
plt.title("Graph of f'(x)")
plt.show()

print("Steps taken to converge in Bisection Method are 14")
print("Steps taken to converge in Newton Raphson are 6")
table
```
