```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x  = np.arange(-2,2,0.01)
y = np.exp(x)

plt.plot(x,y,label= "e^x")
plt.grid(True)

y = x
plt.plot(x,y,label= "x")
plt.grid(True)

x  = np.arange(0.001,2,0.01)
y = np.log(x)
plt.plot(x,y,label= "ln(x)")
plt.grid(True)
plt.legend(loc='best')
plt.show()

x  = np.arange(-4,4,0.01)
a = [1, -1]
for i in a: 
    y = np.exp((i)*x)
    plt.plot(x,y,label= ("e^( "+ str(i) +"x)"))
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

We observe that on plotting the graph of the exponential function, as x tends to infinity, the value of the function also tends to infinity with a monotonously increasing slope which is equal to the function itself.
Also, as the value of x tends to negative infinity the value of the function tends to 0.

In the graph of the linear function, as x tends to infinity, the value of the function also tends to infinity with a slope which is equal 1. As the value of x tends to negative infinity the value of the function tends to negative infinity.

For non-positive values of x, the logarithmic functnion is not defined. On plotting the graph of the logarithmic function, for positive values of x tends to infinity, the value of the function also tends to infinity with a slope which is equal 1/x. As the value of x tends to 0 the value of the function tends to negative infinity. At x=1, value of the function is 0.

We can also see that e^(+-x) are mirror images of each other along the y axis.

```python
x  = np.arange(-np.pi,np.pi,0.01)
k  = [-2,-1,1,2]
for i in k:
    y = np.sin(i*x)
    plt.plot(x,y,label= ("sin( "+ str(i) +"x)"))
    plt.grid(True)
plt.legend(loc='best')
plt.show() 
y = np.sin(x)
plt.plot(x,y,label="sin(x)")
plt.plot(x,y**2,label="(sin(x))^2")
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

In sin(kx) as the value of k increases, we observe that the graph starts shrinking across the x axis and vice versa.

We observe that sin(x)^2 is always positive and its value is always smaller than or equal to that of mod(sin(x)).

```python
y0 = 2;
a = .2;
mu = 5;
x = np.arange(0,10,0.001)
y = y0*(np.exp(-a*(x-mu)**2))
plt.plot(x,y,label="2e^(-0.2(x-5)^2)")
plt.grid(True)
plt.legend(loc='best')
plt.show()
y0 = 1;
a = 1;
mu = [1,2,3,4,5];
x = np.arange(-2,8,0.001)
for i in mu:
    y = y0*(np.exp(-a*(x-i)**2))
    plt.plot(x,y,label="e^((x-"+str(i)+")^2)")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

We observe that while changing the value of mu, the maxima of the graph shifts on the x axis, thus we observe that mu is the mean of the gaussian curve.

```python
y0 = 1;
a = 1;
mu = 0;
x = np.arange(-5,5,0.001)
y1 = np.exp(-(x**2))
y2 = 1/(1+x**2);
plt.plot(x,y1,label="Gaussian Curve")
plt.plot(x,y2,label="Lorentz Curve")
plt.grid(True)
plt.legend(loc="best")
plt.show()
plt.plot(x,y1-y2,label="Error")
plt.grid(True)
plt.legend(loc="best")
plt.show()
```

We observe that the Lorentz function is a close approximation of the Gaussian curve around the value of 0.

```python
x = np.arange(0.0001,2,0.001)
y = x*np.log(x)
plt.plot(x,y,label="xlnx")
plt.grid(True)
plt.legend(loc="best")
plt.show()
```

The value of the derivative of the function near 0 is -1, at very high values of x, the value of the derivative of the function is inf. At large values of x, xlnx can be equated to x.

```python
x = np.arange(-3,3,0.001)
a=[0, 1.5, 3.5,5.5]
for i in a:
    y = -i*x + x**3
    plt.plot(x,y,label=str(-i)+"x + x^3")
plt.legend(loc="best")
plt.grid(True)
plt.show()


x = np.arange(-3,3,0.001)
a=[0, 1.5, 3.5,5.5]
for i in a:
    y = -i + 3*(x**2)
    plt.plot(x,y,label=str(-i)+" + 3x^2")
plt.legend(loc="best")
plt.grid(True)
plt.show()

x = np.arange(-3,3,0.0001)
a=[0, 1.5, 3.5,5.5]
for i in a:
    y = 6*(x)
    plt.plot(x,y,label="6x")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

We observe that the value of a increases, the roots of the function go farther from the origin with one root always being at the origin. The graph of the first derivative goes down as the value of a increases. The value of the second derivative is a straight line and is the same for all values of a.

```python
x = np.arange(-3,3,0.001)
a=[0, 1.5, 3.5,5.5]
for i in a:
    y = -i*(x**2) + (x**4)
    plt.plot(x,y,label=str(-i)+"x^2 + x^4")
plt.legend(loc="best")
plt.grid(True)
plt.show()
x = np.arange(-3,3,0.001)
a=[0, 1.5, 3.5,5.5]
for i in a:
    y = -2*i*(x) + 4*(x**3)
    plt.plot(x,y,label=str(-(2*i))+"x + 4x^3")
plt.legend(loc="best")
plt.grid(True)
plt.show()
x = np.arange(-3,3,0.0001)
a=[0, 1.5, 3.5,5.5]
for i in a:
    y = -2*i + 12*(x**2)
    plt.plot(x,y,label=str(-(2*i))+"+ 12x^2")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

We observe that the value of a increases, the roots of the first derivate of teh function go farther from the origin with one root always being at the origin. The graph of the second derivative goes down as the value of a increases.
