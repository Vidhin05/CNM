```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x0 = 0
y0 = 1
xf = 10
k = 1
plt.figure(figsize= (25,12))
for h in [0.1, 0.05]:
    n = int(10/h+1)
    x = np.linspace(x0, xf + h, n)
    y = np.zeros([n])
    y[0] = y0

    for i in range(1, n) :
        y[i] = y[i-1] + h*(-y[i-1] + 2*np.cos(x[i-1]))                                                                      
    
    plt.subplot(2, 4, k)
    plt.plot(x, y, '.')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("h = "+ str(h))
    plt.plot(x, (np.sin(x) + np.cos(x)))
    
    plt.subplot(2, 4, k + 4)
    plt.plot(x, y - (np.sin(x) + np.cos(x)))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("Error for h = "+ str(h))
    k += 1

for h in [0.1, 0.05]:
    n = int(10/h+1)
    x = np.linspace(x0, xf + h, n)
    y = np.zeros([n])
    y[0] = y0

    for i in range(1, n) :
        y[i] = y[i-1] + h*(-y[i-1] + 2*np.cos(x[i-1])) + ((h*h)/2)*(y[i-1] - 2*np.cos(x[i-1]) - 2*np.sin(x[i-1]))
    
    plt.subplot(2, 4, k)
    plt.plot(x, y, '.')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("h = "+ str(h))
    plt.plot(x, (np.sin(x) + np.cos(x)))
    
    plt.subplot(2, 4, k + 4)
    plt.plot(x, y - (np.sin(x) + np.cos(x)))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.title("Error for h = "+ str(h))
    k += 1
```
