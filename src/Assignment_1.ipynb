```python
i = np.arange(0.001, 1, 0.001)
k = 1
y = -k*((i*(np.log2(i))+((1-i)*np.log2(1-i))))
plt.plot(i, y, label ="<I> vs p")
plt.grid(True)
plt.legend(loc = 'best')
plt.show()
```

##  Q1.b

$I =-k(plog_{2}p+(1-p)(log_{2}(1-p)))$

$p = 0.5 + \epsilon$

$\therefore $ $I =-k((0.5+\epsilon)log_{2}(0.5+\epsilon)+(0.5-\epsilon)(log_{2}(0.5-\epsilon)))$

$\therefore I = -k((0.5 + \epsilon)(\frac{(\ln(1+ 2\epsilon)}{\ln{2}})+(0.5 - \epsilon)(\frac{(\ln(1- 2\epsilon)}{\ln{2}}))$

$\therefore I = -k((0.5 + \epsilon)(\frac{2\epsilon}{\ln2})+(0.5 - \epsilon)(\frac{-2\epsilon}{\ln2})$

$\therefore I = \frac{-4k(\epsilon)^2}{\ln(2)} +k$

$\therefore I = a -b \epsilon^2$ $ where$ $ a = k$ $ and $ $b = \frac{4k}{\ln(2)}$
