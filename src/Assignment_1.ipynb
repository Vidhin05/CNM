##  Q1.b

$I =-k(plog_{2}p+(1-p)(log_{2}(1-p)))$

$p = 0.5 + \epsilon$

$\therefore $ $I =-k((0.5+\epsilon)log_{2}(0.5+\epsilon)+(0.5-\epsilon)(log_{2}(0.5-\epsilon)))$

$\therefore I = -k((0.5 + \epsilon)(\frac{(\ln(1+ 2\epsilon)}{\ln{2}})+(0.5 - \epsilon)(\frac{(\ln(1- 2\epsilon)}{\ln{2}}))$

$\therefore I = -k((0.5 + \epsilon)(\frac{2\epsilon}{\ln2})+(0.5 - \epsilon)(\frac{-2\epsilon}{\ln2})$

$\therefore I = \frac{-4k(\epsilon)^2}{\ln(2)} +k$

$\therefore I = a -b \epsilon^2$ $ where$ $ a = k$ $ and $ $b = \frac{4k}{\ln(2)}$

```python
B = 2
R = np.arange(2,0.7254,-0.0001)
vals1 = []
vals2 = []
x1 = []
x2 = []
for r in R:
  coeff = [r**4, 0, -B*(r**4), -4*(r**2), 3]
  roots = np.roots(coeff)
#   print(r)
  [y1, y2] = np.real(roots[np.imag(roots)==0])
  x1.append((r**-2)*(y1**-1))
  x2.append((r**-2)*(y2**-1))
  vals1.append(y1)
  vals2.append(y2)

vals1 = np.array(vals1)
vals2 = np.array(vals2)
x1 = np.array(x1)
x2 = np.array(x2)
plt.plot(R, vals1)
plt.plot(R, vals2)
plt.plot(R, (x1*((3*x1) - 2))**0.5)
plt.plot(R, (x2*((3*x2) - 2))**0.5)
plt.show()
```
