# Write a code, applying the algorithm of the Newton-Raphson method to determine both the real roots of  $f(x) = x^{6} − x − 1=0.$

```python
table = pd.DataFrame(columns=['n', 'x(n)', 'f(x)', "f'(x)", 'x(n) - x(n-1)', 'x(n+1)', 'x(n+1) - x(n)'])

n    = 0
x_n  = 1.0
f_x  = x_n**6 -x_n -1
dfdx = 6*(x_n**5) - 1
dif1 = 0
x_n1 = x_n - (f_x/dfdx)
dif2 = x_n1 - x_n
table.loc[0] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]

while(1):
  n += 1
  dif1 = x_n1 - x_n
  x_n = x_n1
  f_x  = x_n**6 -x_n -1
  dfdx = 6*(x_n**5) - 1
  x_n1 = x_n - (f_x/dfdx)
  dif2 = x_n1 - x_n
  table.loc[n] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]
  if(abs(dif2) < 0.0001):
    break
    
x = np.arange(-3,3,0.001)
y = x**6 -x -1
plot(x,y,"x^6 -x -1")
x = np.arange(-3,3,0.001)
y = 6*x**5 -1
plot(x,y,"6x^5 - 1")

plt.plot(table['n'], table['x(n+1)'])
plt.title("Convergence of root between 1 and 2")
plt.show()
plt.plot(table["n"], table["f'(x)"])
plt.title("Graph of f'(x)")
plt.show()

print("Steps taken to converge in Bisection Method are 14")
print("Steps taken to converge in Newton Raphson are 6")
table
```

# 4. Find the smallest non-zero positive root of $x = tanx$, with an accuracy of $\epsilon = 0.0001$. Further solve for the root that is closest to $x = 100$.

```python
table = pd.DataFrame(columns=['n', 'x(n)', 'f(x)', "f'(x)", 'x(n) - x(n-1)', 'x(n+1)', 'x(n+1) - x(n)'])

n    = 0
x_n  = 4.5
f_x  = x_n - np.tan(x_n)
dfdx = 1 - (np.cos(x_n)**(-2))
dif1 = 0
x_n1 = x_n - (f_x/dfdx)
dif2 = x_n1 - x_n
table.loc[0] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]

while(1):
  n += 1
  dif1 = x_n1 - x_n
  x_n = x_n1
  f_x  = x_n - np.tan(x_n)
  dfdx = 1 - (np.cos(x_n)**(-2))
  x_n1 = x_n - (f_x/dfdx)
  dif2 = x_n1 - x_n
  table.loc[n] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]
  if(abs(dif2) < 0.0001):
    break
x = np.arange(-10,10,0.1)
y = np.tan(x) - x 
plot(x,y,"tan(x) - x")
plot(x,(np.cos(x))**(-2) -1,"f'x")
plt.plot(table['n'], table['x(n+1)'])
plt.title("Convergence of root between 1 and 2")
plt.show()
plt.plot(table["n"], table["f'(x)"])
plt.title("Graph of f'(x)")
plt.show()
print("Steps taken to converge in Bisection Method are 19")
print("Steps taken to converge in Newton Raphson are 4")

table
```

# The function $y=f(x)=a+x(x−1)^{2}$,with $0≤a≤0.1$. When $a \neq 0$, there is only one real root of f(x) = 0, with the root being negative. Analytically check how many roots are obtained for a = 0, and what is the nature of the roots. Thereafter, using the Newton-Raphson method, test for the convergence towards the negative real root, through a series of suitably chosen a values going right down to a = 0 (the most important case). In every case your initial guess value should be slightly larger than 1, say 1.01, and slightly smaller than 1, say 0.99. For every value of a, starting from both sides of x = 1, check how quickly the convergence happens.

```python
x = np.arange(-0.2, 1.5, 0.001)

d={}
for a in range(0, 12, 2):
  y = (a/100) + x*((x-1)**2)
  plt.plot(x, y, label='a = '+ str(a/100))
  d[a/2] = pd.DataFrame(columns=['n', 'x(n)', 'f(x)', "f'(x)", 'x(n) - x(n-1)', 'x(n+1)', 'x(n+1) - x(n)'])

  n    = 0
  x_n  = 1.01
  f_x  = (a/100) + x_n*((x_n-1)**2)
  dfdx = ((x_n-1)**2) + (2*x_n*(x_n-1))
  dif1 = 0
  x_n1 = x_n - (f_x/dfdx)
  dif2 = x_n1 - x_n
  d[a/2].loc[0] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]

  while(1):
    n += 1
    dif1 = x_n1 - x_n
    x_n = x_n1
    f_x  = (a/100) + x_n*((x_n-1)**2)
    dfdx = ((x_n-1)**2) + (2*x_n*(x_n-1))
    x_n1 = x_n - (f_x/dfdx)
    dif2 = x_n1 - x_n
    d[a/2].loc[n] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]
    if(abs(dif2) < 0.0001):
      break
    
plt.grid(True)
plt.legend()
plt.show()
d[0]
```
