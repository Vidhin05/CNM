```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('precision', 2)
pd.set_option('display.max_rows', 10)
```

###Second-Order RKM

```python
f = lambda x,y: -y + 2*np.cos(x)
x0 = 0
y0 = 1
h = 0.05
n = np.int32(10/h)
x = np.linspace(0,10,(10/h)+1)
y = np.zeros(n+1)
y[0] = y0
for i in range(n):
  y[i+1] = y[i] + (0.5*h)*(f(x[i],y[i]) + f(x[i] + h, y[i] + h*f(x[i],y[i])))
  
arr = []
arr.append(x)
arr.append(y)
arr = np.transpose(arr)
df = pd.DataFrame(data=arr, columns=['x', 'y'])

plt.plot(df['x'], df['y'], label="Second-Order Runge Kutta Method")
sork = df['y']
y = np.zeros(n+1)
y[0] = y0
for i in range(n):
  v1 = f(x[i], y[i])
  v2 = f(x[i]+(h/2), y[i]+((h/2)*v1))
  v3 = f(x[i]+(h/2), y[i]+((h/2)*v2))
  v4 = f(x[i]+h, y[i]+(h*v3))
  y[i+1] = y[i] + (h/6)*(v1 + 2*(v2 + v3) + v4)

arr = []
arr.append(x)
arr.append(y)
arr = np.transpose(arr)
df = pd.DataFrame(data=arr, columns=['x', 'y'])
plt.plot(df['x'], df['y'], label="Fourth-Order Runge Kutta Method")
fork = df['y']

plt.plot(x, np.sin(x) + np.cos(x), label="Analytic Function")
plt.legend()
plt.show()

a = np.sin(x) + np.cos(x)
plt.plot(x, sork-a, label="Error of Second Order Runge Kutta Method")
plt.legend()
plt.show()
plt.plot(x, fork-a, label="Error of Fourth Order Runge Kutta Method")
plt.legend()
plt.show()
```
