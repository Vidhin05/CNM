```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x  = np.arange(-1.5,2.5,0.01)
y = np.exp(x)
plt.plot(x,y)
y1=(1+x)
plt.plot(x,y1)
y2=(1+x+((x**2)/2))
plt.plot(x,y2)
y3=(1+x+((x**2)/2)+((x**3)/6))
plt.plot(x,y3)
plt.grid(True)
plt.show()
plt.plot(x,y-y1)
plt.plot(x,y-y2)
plt.plot(x,y-y3)
plt.grid(True)
plt.show()
```

We observe that the odd orders of the taylor polynomial expansions are on one side of the graph whereas the even orders of the taylor series expansion are on the other side.
As the order increases we observe that the estimation approaches the real value of the function.

```python
x  = np.arange(0.0001,3,0.01)
y = np.log(x)
plt.plot(x,y)
plt.grid(True)
y1 = x-1
y2 = y1-(((x-1)**2)/2)
y3 = y2 + ((x-1)**3)/math.factorial(3)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid(True)
plt.show()
plt.plot(x,(y-y1))
plt.plot(x,(y-y2))
plt.plot(x,(y-y3))
plt.grid(True)
plt.show()
```

```python
x = np.arange(-np.pi,np.pi,0.01) 
y = np.sin(x)
plt.plot(x,y)
plt.grid(True)
y1 = x
y2 = y1 - x**3/math.factorial(3)
y3 = y2 - x**5/math.factorial(5)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
plt.plot(x,(y-y1))
plt.plot(x,(y-y2))
plt.plot(x,(y-y3))
plt.grid(True)
plt.show()
```

We observe that the odd orders of the taylor polynomial expansions are on one side of the graph whereas the even orders of the taylor series expansion are on the other side.
As the order increases we observe that the estimation approaches the real value of the function.

```python
x = np.arange(-np.pi,np.pi,0.01) 
y = np.cos(x)
plt.plot(x,y)
plt.grid(True)
y1 = x*0 + 1
y2 = y1 - x**2/math.factorial(2)
y3 = y2 - x**4/math.factorial(4)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid(True)
plt.show()
plt.plot(x,(y-y1))
plt.plot(x,(y-y2))
plt.plot(x,(y-y3))
plt.grid(True)
plt.show()
```

As the order increases we observe that the estimation approaches the real value of the function.
